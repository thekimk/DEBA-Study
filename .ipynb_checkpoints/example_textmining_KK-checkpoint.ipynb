```python
import re
import string
import kss    # ë¬¸ì¥ë¶„ë¦¬
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords    # ì˜ì–´ ë¶ˆìš©ì–´ ì œê±°
## í•œêµ­ì–´
from soynlp.normalizer import *
from soynlp.word import WordExtractor
from soynlp.noun import LRNounExtractor, LRNounExtractor_v2, NewsNounExtractor
from soynlp.tokenizer import LTokenizer
```

```python
def text_preprocessor(text, bracket_content=False):
    # í•œê¸€ ë§ì¶¤ë²•ê³¼ ë„ì–´ì“°ê¸° ì²´í¬ (PyKoSpacing, Py-Hanspell)
    # html íƒœê·¸ ì œê±°í•˜ê¸°
    text_new = re.sub(r'<[^>]+>', '', text)
    # ê´„í˜¸ì™€ ë‚´ë¶€ë¬¸ì ì œê±°í•˜ê¸°
    if bracket_content:
        text_new = re.sub(r'\([^)]*\)', '', text_new)
        text_new = re.sub(r'\[[^)]*\]', '', text_new)
        text_new = re.sub(r'\<[^)]*\>', '', text_new)
        text_new = re.sub(r'\{[^)]*\}', '', text_new)
    # ì˜ì–´(ì†Œë¬¸ìí™”), í•œê¸€, ìˆ«ìë§Œ ë‚¨ê¸°ê³  ì œê±°í•˜ê¸°
    text_new = re.sub('[^ A-Za-z0-9ê°€-í£]', '', text_new.lower())
    # í•œê¸€ ììŒê³¼ ëª¨ìŒ ì œê±°í•˜ê¸°
    text_new = re.sub('([ã„±-ã…ã…-ã…£]+)', '', text_new)
    # ìˆ«ì ì œê±°í•˜ê¸°
    text_new = re.sub(r'\d+', '', text_new)
    # í•œê¸€ ë° ì˜ì–´ stopwords ì œê±°í•˜ê¸°
    # ë¬¸ì¥êµ¬ë‘ì  ë° ì–‘ìª½ê³µë°± ì œê±°í•˜ê¸°
    translator = str.maketrans('', '', string.punctuation)
    text_new = text_new.strip().translate(translator)
#     # ê³µë°± ì œê±°í•˜ê¸°
#     text_new = re.sub(r'\s+', '', text_new)
    # 2ê°œ ì´ìƒì˜ ë°˜ë³µê¸€ì ì¤„ì´ê¸°
    text_new = ' '.join([emoticon_normalize(word, num_repeats=2) for word in text_new.split(' ')])
    text_new = ' '.join([repeat_normalize(word, num_repeats=2) for word in text_new.split(' ')])
    
    return text_new
    
df = df_news.copy()

# ì „ì²˜ë¦¬
df.drop_duplicates(inplace=True, ignore_index=True)
df.Category = df.Category.apply(lambda x: text_preprocessor(x, bracket_content=False))
df.Title = df.Title.apply(lambda x: text_preprocessor(x, bracket_content=True))
df.Content = df.Content.apply(lambda x: text_preprocessor(x, bracket_content=False))
df
# ë¬¸ì„œ ìš”ì•½
```

```python
def preprocessing_nounextract(df_series):
    # ë‹¨ì–´ ì¶”ì¶œê¸°
    ## cohesion/branching entropy/accessor varietyê°’ì´ í° ê²½ìš° í•˜ë‚˜ì˜ ë‹¨ì–´ì¼ ê°€ëŠ¥ì„± ë†’ìŒ
    word_extractor = WordExtractor()
    word_extractor.train(list(df_series.values))
    word_score = word_extractor.extract()
    ## cohesion_forward*right_branching_entropy ë¡œ scoring
    words = {word:score.cohesion_forward * math.exp(score.right_branching_entropy) for word, score in word_score.items()}
    nouns_WE = sorted(words.items(), key=lambda x:x[1], reverse=True)

    # ë‹¨ì–´ì¶”ì¶œê¸°
    noun_extractor = LRNounExtractor_v2(verbose=False, extract_compound=True)
    noun_extractor.train(list(df_series.values))
    noun_score = noun_extractor.extract()
    ## unique ë‹¨ì–´ë“¤ë¡œ í•„í„°
    unique_nouns = list(set([words for word_tuple in noun_extractor._compounds_components.values() for words in word_tuple]))
    nouns = {noun:noun_score[noun] for noun in unique_nouns if noun in noun_score.keys()}
    ## ë¹ˆë„ìˆ˜*ëª…ì‚¬ì ìˆ˜ ë¡œ scoring
    nouns = {noun:score.frequency * score.score for noun, score in nouns.items()}
    nouns_LRE = sorted(nouns.items(), key=lambda x:x[1], reverse=True)
    
    # ë‹¨ì–´ì¶”ì¶œê¸°
    noun_extractor = NewsNounExtractor(verbose=False)
    noun_extractor.train(list(df_series.values))
    noun_score = noun_extractor.extract()
    ## ë¹ˆë„ìˆ˜*ëª…ì‚¬ì ìˆ˜ ë¡œ scoring
    nouns = {noun:score.frequency * score.score for noun, score in noun_score.items()}
    nouns_NE = sorted(nouns.items(), key=lambda x:x[1], reverse=True)

    return nouns_WE, nouns_LRE, nouns_NE

preprocessing_nounextract(df.Title)
```
